```python
#plotting cdf from pmf

import numpy as np
import matplotlib.pyplot as plt

#pmf and corresponding x values
p = np.array([0.25, 0.5, 0.25])
x = np.array([0, 1, 4])

#cumulative sum
F = np.cumsum(p)
plt.stem(x,p); plt.show()
plt.step(x,F); plt.show()

#expectation
EX = np.sum(p*x)
print("Expectation:",EX)

#variance
VarX = np.sum(p*(x-EX)*(x-EX))
print("Variance:", VarX)
```

```python
# Python code to compute pmf and cdf of binomial R.V
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
p = 0.5
n = 10
rv = stats.binom(n,p)
x = np.arange(11)
F = rv.cdf(x)
p=rv.pmf(x);

plt.figure(1); plt.plot(x, F, 'bo-', ms=10);
plt.xlabel("x")
plt.ylabel("CDF")
plt.figure(2); plt.plot(x, p, 'bo')
plt.vlines(x, 0, p, colors='b', lw=5, alpha=0.5);
plt.xlabel("x")
plt.ylabel("PMF")
```

```python
#PDF of Binomial for various n, p

# Python code to compute pmf and cdf of binomial R.V
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

n=100
p0=0.1;
p1=0.5;
p2=0.9;

rv0 = stats.binom(n,p0)
rv1 = stats.binom(n,p1)
rv2 = stats.binom(n,p2)

x = np.arange(n+1);
p0=rv0.pmf(x);
p1=rv1.pmf(x);
p2=rv2.pmf(x);

plt.figure(1); plt.plot(x, p0, 'bo', ms=5);
plt.vlines(x, 0, p0, colors='b', lw=2, alpha=0.5);
plt.plot(x, p1, 'yo', ms=5);
plt.vlines(x, 0, p1, colors='y', lw=2, alpha=0.5);
plt.plot(x, p2, 'ro', ms=5);
plt.vlines(x, 0, p2, colors='r', lw=2, alpha=0.5);

plt.xlabel("x, n=100")
plt.ylabel("PMF")

print(p2[89])
print(p2[90])
print(p2[91])

n1=120;
n2=150;
rv3 = stats.binom(n,0.3)
rv4 = stats.binom(n1,0.3)
rv5 = stats.binom(n2,0.3)

x = np.arange(n+1);
p3=rv3.pmf(x);
p4=rv4.pmf(x);
p5=rv5.pmf(x);

plt.figure(2); plt.plot(x, p3, 'bo', ms=5);
plt.vlines(x, 0, p3, colors='b', lw=2, alpha=0.5);
plt.plot(x, p4, 'yo', ms=5);
plt.vlines(x, 0, p4, colors='y', lw=2, alpha=0.5);
plt.plot(x, p5, 'ro', ms=5);
plt.vlines(x, 0, p5, colors='r', lw=2, alpha=0.5);

plt.xlabel("x, p=0.3")
plt.ylabel("PMF")

plt.figure(3); plt.plot(x/n, p3, 'bo', ms=5);
plt.plot(x/n1, p4, 'yo', ms=5);
plt.plot(x/n2, p5, 'ro', ms=5);
plt.vlines(x/n, 0, p3, colors='b', lw=2, alpha=0.5);
plt.vlines(x/n1, 0, p4, colors='y', lw=2, alpha=0.5);
plt.vlines(x/n2, 0, p5, colors='r', lw=2, alpha=0.5);


plt.xlabel("x/n, p=0.3")
plt.ylabel("PMF")
plt.legend(['n=100','n=120','n=150'])
```

```python
#Poisson RV
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

lam = 10;
x = np.arange(100);

rv0 = stats.poisson(lam)
p0 = rv0.pmf(x);


plt.figure(1); plt.plot(x, p0, 'bo', ms=5);
plt.vlines(x, 0, p0, colors='b', lw=2, alpha=0.5);
```

```python
#Binomial Approximation to Poisson

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

lam = 10;
n = 100;
p=0.1;

x = np.arange(100);

rv0 = stats.poisson(lam)
p0 = rv0.pmf(x);
rv1 = stats.binom(n, p)
p1 = rv1.pmf(x);


plt.figure(1); plt.plot(x, p0, 'bo', ms=5);
plt.plot(x, p1, 'go', ms=5);
plt.vlines(x, 0, p0, colors='b', lw=2, alpha=0.5);
plt.vlines(x, 0, p1, colors='g', lw=2, alpha=0.5);
plt.legend(["Poisson $\lambda = 10$", "Binomial, n=100, p=0.1"])
```

```python
#Geometric R.V
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

p = 0.3;
x = np.arange(100);

rv0 = stats.geom(p)
p0 = rv0.pmf(x);
f0 = rv0.cdf(x);


plt.figure(1); plt.plot(x, p0, 'bo', ms=5);
plt.vlines(x, 0, p0, colors='b', lw=2, alpha=0.5);
plt.ylabel('PMF');
plt.xlabel('x');

plt.figure(2);plt.step(x, f0, 'b', where="post");
plt.ylabel('CDF');
plt.xlabel('x');
```

```python
# Python code to plot the exponential PDF
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

lambd1 = 1/2
lambd2 = 1/5
x = np.linspace(0,1,1000)
f1 = stats.expon.pdf(x,scale=lambd1)
f2 = stats.expon.pdf(x,scale=lambd2)
plt.plot(x, f1)
plt.plot(x, f2)
plt.legend([r'$\lambda=2$',r'$\lambda=5$'])
```

```python
#Exponential R.V and Approximation to Geometric R.V

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math


lambd1 = 1/2

x=np.linspace(0,3.5,100)
F1 = stats.expon.cdf(x,scale=lambd1)

plt.figure(1); plt.plot(x, F1)

x_geom=np.arange(20);
print(x_geom)
p = 0.3;
rv0 = stats.geom(p)
F0 = rv0.cdf(x_geom);
delta = (math.log(1/(1-p)))*lambd1;

plt.step(x_geom*delta, F0, where="post");
plt.ylabel('CDF');
plt.xlabel('x');
plt.legend(['Exponential','Geometric'])
```

```python
#Normal R. V

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math


mu=0; sigma1=2; sigma2=4;
x=np.linspace(-10,10,1000)
f1 = stats.norm.pdf(x,mu, sigma1)
f2 = stats.norm.pdf(x,mu, sigma2)

plt.figure(1); plt.plot(x, f1)
plt.plot(x, f2)
plt.legend([r"$\sigma=2$", r"$\sigma=4$"])

F1 = stats.norm.cdf(x,mu, sigma1)
F2 = stats.norm.cdf(x,mu, sigma2)

plt.figure(2); plt.plot(x, F1)
plt.plot(x, F2)
plt.legend([r"$\sigma=2$", r"$\sigma=4$"])
```

```python
#Distribution of Sample mean and its approximation using Normal distribution
# Sample mean of Bernoulli R.Vs or Poisson R.Vs or uniform RV
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

p=0.3; n=1;
ns = 5000;
m=10;

X = np.random.binomial(n, p, size=ns*m)

X=X.reshape(m, ns);
S_samples = (np.sum(X, axis=0)-m*p)/math.sqrt(m) #ns samples of sample means

plt.hist(S_samples,bins='auto');

m=100; ns=50000

X = np.random.binomial(n, p, size=ns*m)

X=X.reshape(m, ns);
S_samples = (np.sum(X, axis=0)-m*p)/math.sqrt(m) #ns samples of sample means

plt.hist(S_samples,bins='auto');

mu=0; sigma1=p*(1-p);
x=np.linspace(-5,5,1000)
F1 = stats.norm.pdf(x,mu, sigma1)
plt.figure(2); 
plt.plot(x,F1*ns,'g')
```
